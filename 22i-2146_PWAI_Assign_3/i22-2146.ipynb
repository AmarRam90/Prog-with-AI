```python
excel_file = 'C:/Users/Arham Shakeel/Desktop/FastUniversity/3rdSemester/Prog with AI/22i-2146_PWAI_Assign_3/timetable.xlsx'
df = pd.read_excel(excel_file, engine='openpyxl')
```

```python
workbook = load_workbook(excel_file, data_only=True)

# Select the sheet you want to work with (e.g., the first sheet)
sheet = workbook['Monday']
# Get the dimensions of the sheet
max_row = sheet.max_row
max_col = sheet.max_column

batch_cell_value = 0
stored_cell_value = 0
search_value = input("Enter Batch")
for index,row in df.iterrows():
     for column in df.columns:
        batch_cell_value = row[column]
        if batch_cell_value == search_value:
            stored_cell_value = row[column]

# Ask the user for a cell value

# Create a function to get the color code of a cell
def get_color_code(cell):
    return cell.fill.start_color.rgb if cell.fill.start_color.rgb else None

# Find the color code of the user-specified cell
target_color = None
for row_index in range(1, max_row + 1):
    for col_index in range(1, max_col + 1):
        cell = sheet.cell(row=row_index, column=col_index)
        if cell.value == stored_cell_value:
            target_color = get_color_code(cell)
            break
    if target_color:
        break

if target_color:
    print(f"The Monday Classes for '{stored_cell_value}'are ")
else:
    print(f"Cell with value '{stored_cell_value}' not found in the sheet.")
    exit()

# Create a dictionary to store cells by color code
similar_color_cells = {}
col_indexes_of_cell_values = []
row_indexes_of_cell_values = []

# Iterate over all rows and columns to find cells with similar color codes
for row_index in range(1, max_row + 1):
    for col_index in range(1, max_col + 1):
        cell = sheet.cell(row=row_index, column=col_index)
        #print(f"Here is the check {col_index}")
        cell_color = get_color_code(cell)

        # Check if the cell has a similar color code
        if cell_color == target_color:
            # Store the cell in the dictionary
            #col_indexes_of_cell_values.append(col_index)
            if target_color in similar_color_cells:
                similar_color_cells[target_color].append(cell.value)
            else:
                similar_color_cells[target_color] = [cell.value]


#for value in similar_color_cells[target_color]:
#    print(value)

#for value in col_indexes_of_cell_values:
 #   print(f'Here is the check {}')

# Display the values of cells with similar color codes
for color_code, cell_values in similar_color_cells.items():
    for cell_value in cell_values:
        #for this_cell_row_index in row_indexes_of_cell_values:
           #print(f'Subject and Section: {cell_value}, Class Room: ')
           user_classes = cell_value
           for index,row in df_1.iterrows():
                for column in df_1.columns:
                    temp_cell_value = row[column]
                    if temp_cell_value == user_classes:
                       print(f':Class: {temp_cell_value} , Timings:{column}, Room:{index} ')
```
